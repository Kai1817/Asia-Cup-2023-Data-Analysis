### Lets look at the number of matches won by the teams in Asia cup 2023.

```python
win = df['Winner'].value_counts()
plt.figure(figsize=(7,5))
sns.countplot(x='Winner',data=df,order=win.index,palette='magma')
plt.title("Number of Matches won by the Teams");
```

**India** lifts the AsiaCup 2023 with a win of 4 matches all over,**Srilanka** too won 4 matches and stayed as Runner's up team.
Both **Pakistan** and **Bangladesh** won 2 matches only whereas,**Afghanistan** and **Nepal** did'nt won a single match in the AsiaCup 2023.The group stage match between **India and Pakistan** ends with no result resulting in the outcome of 12 out of 13 matches only.

### Number of matches won by the teams batting first or second.

```python
u =  df['Won by'].value_counts()
plt.figure(figsize=(4,4))
sns.countplot(x='Won by',data=df,order=u.index,palette='viridis_r')
plt.title("Teams won by batting first or second");
```

Teams **batting first won 8 times** where as teams **chasing won 4 times**.

### Decisions taken after winning the toss.

```python
l = df['Toss decision'].value_counts()
plt.figure(figsize=(4,4))
sns.countplot(x='Toss decision',data=df,order=l.index,palette='cividis')
plt.title("Team decision after winning the toss");
```

After winning the toss **9 times teams have decided to bat first** and **4 times teams have decided to field first**.

### Top scorers in the Asiacup 2023

```python
p = df.groupby(['Top scorer'])['Highest score'].value_counts()
p.sort_values(ascending = False)
p.to_frame()
```

**Shubman Gill, Rohit Sharma and Kusal Mendis** remained the top scorers multiple times.

```python
p.plot(kind='barh',figsize=(7,5),color='tomato')
plt.ylabel('Top Scorers')
plt.xlabel('count')
plt.title('Top scorers of the tournament');
```

### Top scorers who have scored a Hundred in the AsiaCup.

```python
y = df[df['Highest score'] >= 100].groupby('Top scorer').size()
y.to_frame()
```

### Number of player of the match award.

```python
t = df['Player of the match'].value_counts()
plt.figure(figsize=(9,5))
sns.barplot(x=t.values,y=t.index,palette='viridis')
plt.ylabel('Players')
plt.xlabel('count')
plt.title("Player of the Match");
```

Only **Kusal Mendis** is the player who have got 2 player of match awards the rest of the players have got only 1.

### Balls taken by the Top scorers to score a hundred.

```python
k = df.groupby(['Top scorer','Balls faced'])['Highest score'].max()
l = k[k>=100]
l.to_frame()
```

```python
l.plot(kind='bar',figsize=(11,4),color='olivedrab')
plt.xlabel('Top Scorer with balls faced')
plt.ylabel('scores above 100')
plt.title('Balls taken by the Top scorers to score a hundred ')
plt.xticks(rotation=0);
```

**Virat Kohli** is the only batsman who had scored a hundred in less than 100 balls, enhancing his strike rate and average whereas the rest of the batsman have taken more than 100 balls to score a hundred.

### Venues of Asiacup 2023

```python
print(df['Venue'].unique())
```

### **Venue that conducted most number of matches**

```python
r = df['Venue'].value_counts()
plt.figure(figsize=(8, 5))  
sns.barplot(x=r.values,y=r.index, palette='cividis_r')
plt.xlabel('Number of Matches')  
plt.ylabel('Venue')
plt.title('Count of Matches Hosted by different Venues');
```

Hence **R.Premadasa Stadium** had hosted the most number of matches in the Asiacup 2023.

### Average scores while batting first and batting second at different venues

```python
h = df.groupby(['Venue'])['First innings score'].mean()
i = df.groupby(['Venue'])['Second innings score'].mean()
stadium = pd.concat([h,i],axis=1)
stadium
```

```python
stadium.plot(kind='barh',figsize=(9,5))
plt.xlabel('Scores')
plt.title('Venues with First & Second innings average Scores');
```

### Average fall of wickets in first and second innings at different venues

```python
v = df.groupby(['Venue'])['First innings wickets'].mean()
m = df.groupby(['Venue'])['Second innings wickets'].mean()
stad_wickets = pd.concat([v,m],axis=1)
stad_wickets
```

```python
stad_wickets.plot(kind='barh',figsize=(9,5))
plt.xlabel('Wickets')
plt.title('Venues with First & Second innings average fall of Wickets');
```

- So definitely Batting first in all the venues had been productive for the teams,Specialy the **Multan Cricket stadium** which seems to be the best suited for a team batting first.
- Teams batting first had won most of the matches comparitively in all the venues. 
- Teams bowling first had more advantage comparatively to the teams who bowled second except in the venue **Multan Cricket Stadium**,it supports team bowling second more than the teams bowling first.
- Specially the **Pallekele International Cricket Stadium** is best suited for defending a total.
